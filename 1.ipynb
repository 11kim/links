```python
VOCALS = ['а', 'я', 'э', 'е', 'у', 'ю', 'ы', 'и', 'о', 'ё']

def syll(word):
    count = 0
    for s in word:
        if s in VOCALS:
            count += 1
    return count
```

```python
from russtress import Accent

def accents(line):
    accent = Accent()
    accented_text = accent.put_stress(line)
    return accented_text
```

```python
fin = open('text.txt', 'r')
text = fin.readlines()
fin.close()
for i in range(len(text)):
    text[i] = text[i].rstrip()
    for s in ('.', ',', ';', ':', '«', '»', '?', '!', '―'):
        text[i] = text[i].replace(s, '')
    text[i] = text[i].lower()
    text[i] = text[i].split()
print(text)
```

```python
res = []
for line in text:
    for word in line:
        #print(word, syll(word))
        if syll(word) >= 4:
            #print(word, syll(word))
            res.append(word)
print(len(res))
```

```python
VOCALS = ['а', 'я', 'э', 'е', 'у', 'ю', 'ы', 'и', 'о', 'ё']
def acc(acc_word):
    #word = accents(word)
    word = acc_word
    print(word)
    count = 1
    word = word[::-1]
    for s in word:
        if s in VOCALS:
            count += 1
        if s == "'":
            break
    return count
```

```python
need = []
need2 = []
res_acc = []
for word in res:
    print(word)
    word = accents(word)
    res_acc.append(word)
    num = acc(word)
    if num == 4:
        need.append(word)
    if num >= 4:
        need2.append(word)

print(need)
print(need2)
```
